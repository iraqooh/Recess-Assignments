```python
# Loading the pandas module
import pandas as pd
```

### Series

#### Question 1

```python
series_A = pd.Series([4, 65, 436, 3, 9])
series_B = pd.Series([7, 0, 3, 897, 9])

cmp = series_A == series_B
for index in range(cmp.size):
    print(f'Element {series_A[index]} in series A is {"equal" if cmp[index] else "not equal"} to {series_B[index]} in series B.')
```

#### Question 2

```python
series_A = pd.Series([2, 4, 6, 8, 14])
series_B = pd.Series([1, 3, 5, 7, 9])

series_sum = series_A + series_B
difference = series_A - series_B
product = series_A * series_B
quotient = series_A / series_B

print(f'Sum = {list(series_sum)}')
print(f'Difference = {list(difference)}')
print(f'Product of = {list(product)}')
print(f'Quotient of = {list(map(lambda x: float(format(x, ".2f")), list(quotient)))}')
```

#### Question 5

```python
# Loading dataset
df = pd.read_csv('BSSE 2301 EXERCISE 2/AfricaCupofNationsMatches.csv')

# Extracting subset of dataframe
home = df['HomeTeamGoals']

# Checking data type in the HomeTeamGoals Column
if(home.dtype != 'float64'):
    pd.to_numeric(home, errors='coerce')
    
# Setting null cells in HomeTeamGoals Column to 0
df.loc[home.isna(), 'HomeTeamGoals'] = 0

# Getting the mode of series
moden = home.mode()[0]

# Replacing every non-modal value with 'Other'
home[home != moden] = 'Other'

# Displaying result
print(f'Most frequent goal by a home team is {round(moden)}')
```
