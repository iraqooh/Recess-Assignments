```python
# 'SCHOOL' column has a single empty cell
# Getting entry with null value in 'SCHOOL' column
df[df['SCHOOL'].isna()]
```

```python
# Setting value of empty cell in 'SCHOOL' column
df.loc[df['SCHOOL'].isna(), 'SCHOOL'] = 'UNKNOWN SECONDARY SCHOOL IN HOIMA'
```

```python
# 'TOTAL CANDIDATES' has 271 NaNs but 'FEMALE CANDIDATES' & 'MALE CANDIDATES' columns are all filled
# 'TOTAL CANDIDATES' = 'FEMALE CANDIDATES' + 'MALE CANDIDATES'
# Calculating 'TOTAL CANDIDATES' for each NaN entry
df.loc[df[cols[3]].isna(), cols[3]] = df.loc[df[cols[3]].isna(), cols[18]] + df.loc[df[cols[3]].isna(), cols[33]]
```

```python
# 'TOTAL DIV x' columns have missing values, but since their corresponding '% DIV x' are all available
# 'TOTAL DIV x' = 'TOTAL CANDIDATES' * '% DIV x'/100
def set_division_totals(df, start, indices, static_total=False):
    current_index = start
    for index in indices:
        if static_total:
            df.loc[(df[cols[index]].isna()) & ~(df[cols[index+1]].isna()), cols[index]] = df.loc[(df[cols[index]].isna()), cols[index+1]] * df.loc[(df[cols[index]].isna()), cols[current_index]] / 100
        else:
            df.loc[(df[cols[index]].isna()) & ~(df[cols[index+1]].isna()), cols[index]] = df.loc[(df[cols[index]].isna()), cols[index+1]] * df.loc[(df[cols[index]].isna()), cols[current_index]] / 100
            current_index += 2
```

```python
set_division_totals(df, 3, range(4, 17, 2), True)
set_division_totals(df, 4, range(19, 32, 2), False)
set_division_totals(df, 4, range(34, 47, 2), False)
```

```python
# Analyzing columns with null values
f = list(df.isnull().sum())
for i in range(len(cols)):
    print(f'{cols[i]:20} => {f[i]:5} at {i}')
```

```python
# 'GENDER % div' columns have null values, which can be rectified using the following formula
# 'GENDER % div' = 'GENDER TOTAL DIVx' * 100 / 'TOTAL DIV x'
def set_division_percentages(df, indices, total_start_index):
    current_index = total_start_index
    for index in indices:
        df.loc[df[cols[index]].isna(), cols[index]] = df.loc[df[cols[index]].isna(), cols[index-1]] * 100 / df.loc[df[cols[index]].isna(), cols[current_index]]
        df.loc[df[cols[index]].isna(), cols[index]] = 0
        current_index += 2
```

```python
set_division_percentages(df, range(20, 33, 2), 4)
set_division_percentages(df, range(35, 48, 2), 4)
```

```python
# Summing up data in all numeric columns
sums = df.iloc[:, 3:].sum()
for i in range(len(sums)):
    print(f'{cols[i+3]:20} => {sums[i]:10.3f} at {i}')
```

```python
divisions, labels = [], []
for index in range(1, 14, 2):
    divisions.append(sums[index])
    labels.append(cols[index+3])

plt.pie(divisions, labels=labels, autopct='%1.1f%%')
plt.title('Number of Candidates per Division')

# Display the chart
plt.show()
```

```python
# Gender distribution plots
def plot_gender(start, *titles):
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12, 5))
    index = start
    for row, col in enumerate(axes.flatten()):
        if row == 8: break
        if row < 9:
            col.pie([sums[index], sums[index+15]], labels=[cols[index+3], cols[index+18]], autopct='%1.1f%%')
            col.set_title('Division ' + titles[row])
            index = index+2 if index>15 else index+1
        else: col.axis('off')
    fig.tight_layout()
    plt.show()
            

plot_gender(15, 'All', '1', '2', '3', '4', '7', '9', 'X')
```
