## Replacing Null Values with Default Values

```python
df['What is your gender?'] = df['What is your gender?'].fillna('Female')
df['Age'] = df['Age'].fillna(23)
df['What is your emotion?'] = df['What is your emotion?'].fillna('Abstract')
df['Year of birth'] = df['Year of birth'].fillna('2000')
df['What is your CGPA?'] = df['What is your CGPA?'].fillna('3.75')
df['Do you have a Girl Friend or BOY Friend'] = df['Do you have a Girl Friend or BOY Friend'].fillna('No')
df['Do you have Depression?'] = df['Do you have Depression?'].fillna('No')
df['Do you have Anxiety?'] = df['Do you have Anxiety?'].fillna('Yes')
df['Do you have Panic attack?'] = df['Do you have Panic attack?'].fillna('No')
df['Did you have seek any specialist for a treatment? '] = df['Did you have seek any specialist for a treatment? '].fillna('No')
df.isnull().sum()
```
